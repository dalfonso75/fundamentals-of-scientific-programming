### **Ejercicio 1: Estructuras Básicas de Datos con Python.**

##### a. Defina en una variable el diccionario que creó en el Ejercicio 3 de la Tarea 0 - Reconocimiento de las herramientas y programación básica con Python.

```python
# Diccionario que almacena información sobre personas.
# Cada clave (key) representa el nombre de una persona.
# Cada valor (value) es una tupla con la siguiente estructura:
#
# @key {str} Nombre de la persona
# @value {tuple} Contiene los detalles de la persona:
#     - {int} Edad en años
#     - {float} Peso en kilogramos
#     - {float} Estatura en metros
#     - {bool} Estudia: True si estudia, False si no estudia
#
# Ejemplo:
# "Paula Andrea": (27, 55, 1.55, True)
#     - 27 años
#     - 55 kg
#     - 1.55 m
#     - Estudia: Sí (True)

people = {
    "Paula Andrea": (27, 55, 1.55, True),
    "Elkin Dadey": (30, 60, 1.65, True),
    "Kirley Hernan": (44, 70, 1.76, False),
    "Alveiro Alfonso": (40, 72, 1.67, False),
    "Tulia Ines": (68, 65, 1.60, False),
    "Adis Yamile": (40, 70, 1.60, False),
    "Siervo Hernan": (72, 78, 1.70, False),
    "Diego Alejandro": (26, 78, 1.72, True),
}

print(people)
```

##### b. Implemente ciclos (for o while) para extraer la información del diccionario creado anteriormente y generar las siguientes cuatro (4) listas: Edades, Pesos, Estaturas y Estudia.

```python
# Inicializar listas vacías.
ages = []
weights = []
heights = []
studying = []

# Iterar sobre los valores del diccionario `people`
for data in people.values():
    # Desempaquetar los datos de cada persona
    age, weight, height, is_studying = data

    # Agregar los datos a las listas correspondientes
    ages.append(age)
    weights.append(weight)
    heights.append(height)
    studying.append(is_studying)

# Mostrar las listas generadas
print("Ages:", ages)
print("Weights:", weights)
print("Heights:", heights)
print("Studying:", studying)
```

##### c. Defina una función que reciba como parámetros una lista y regrese (return) el promedio de dicha lista. Utilice dicha función para imprimir (print) el promedio de las listas del inciso anterior

```python
def calculate_average(values):
    """
    Helper para calcular el promedio de una lista de valores numéricos, incluidos números enteros, flotantes y booleanos.
    @param {list} values - Una lista de valores (int, float, or bool).
    @return {float} - El promedio de los valores de la lista.
    """
    # Validar que la lista no esté vacía para evitar división por cero
    if not values:
        return 0
    # Calcular y devolver el promedio
    return sum(values) / len(values)


# Calcular y mostrar el promedio de cada lista
print("Average Age:", calculate_average(ages))
print("Average Weight:", calculate_average(weights))
print("Average Height:", calculate_average(heights))
print(
    "Average Studying:", calculate_average(studying)
)  # Promedio de valores True/False
```

##### e. Programe una función que extraiga todas las listas de los encabezados de columnas que se encuentran en un Panel y úsela para extraer las listas de columnas del Panel del punto anterior. Imprima los resultados de forma que se diferencien los encabezados entre los dataframes en el Panel

```python
def extract_column_headers(panel):
    """
    Hekper para extraer los encabezados de columnas de cada DataFrame en el panel.
    @param {dict} panel - Un diccionario que contiene DataFrames como valores.
    @return {dict} - Un diccionario con las mismas claves que el panel y listas de los encabezados de columnas como valores.
    """
    headers = {}
    for name, df in panel.items():
        # Extraer los nombres de las columnas y almacenarlos en el diccionario
        headers[name] = list(df.columns)
    return headers

# Usar la función para extraer las listas de encabezados de los DataFrames en el panel
column_headers = extract_column_headers(panel)

# Imprimir los encabezados de cada DataFrame en el panel
for name, headers in column_headers.items():
    print(f"\nEncabezados de Columnas {name}:\n{headers}\n")
```

##### f. Programe una función que imprima la cantidad de elementos que tienen las bases de datos en un Panel y úsela con el Panel del punto anterior. Imprima los resultados de forma que se diferencien los tamaños de los diferentes dataframes en el Panel.


```python
def print_dataframe_sizes(panel):
    """
    Helper para imprimir la cantidad de elementos (filas, columnas y total de elementos) de cada DataFrame en el panel.
    @param {dict} panel - Un diccionario que contiene DataFrames como valores.
    """
    for name, df in panel.items():
        rows, columns = df.shape
        total_elements = df.size  # Total de elementos (filas * columnas)
        print(f"\nTamaño del DataFrame '{name}':")
        print(f"Filas: {rows}, Columnas: {columns}")
        print(f"Total de elementos: {total_elements}")

print_dataframe_sizes(panel)
```

### **Ejercicio 3:  Gráficos científicos.**
Desarrolle códigos para generar y visualizar los siguientes gráficos:

##### Gráfico 1: En una misma imagen o visualización (en el plano de coordenadas), presente los siguientes objetos:

- Mostrar los puntos de dispersión del emparejamiento las siguientes listas: 
  - \(x = [13, 14, 15, 16, 17, 18]\)
  - \(y = [16, 24, 45, 57, 75, 102]\)
- Mostrar la función lineal \(f(x) = 17x - 210.3333\) en el dominio \([13, 18]\).
- Mostrar la función exponencial \(g(x) = 0.1452 e^{(0.369x)}\) en el dominio \([13, 18]\).
- Mostrar la función logarítmica \(h(x) = 259.49 \ln(x) - 656.46\) en el dominio \([13, 18]\).
- Los objetos mostrados deben tener colores diferentes.
- La línea de la función logarítmica debe ser punteada.
- En los ejes deben visualizarse los números.
- Asigne un nombre (label) a cada objeto e imprima la leyenda que los identifique en la gráfica.
- Asigne nombres a los ejes de la gráfica.
- Asigne un título a la gráfica.
- Tenga en cuenta que los nombres que decida utilizar le serán de ayuda para presentar las conclusiones que usted obtenga del ejercicio.


```python
import numpy as np
import matplotlib.pyplot as plt

# Puntos de dispersión
x = np.array([13, 14, 15, 16, 17, 18])
y = np.array([16, 24, 45, 57, 75, 102])

# Funciones
def f(x):
    """
    Calcula el valor de la función lineal f(x) = 17x - 210.3333.
    @param {float or np.array} x - Valor(es) de x donde se evaluará la función.
    @return {float or np.array} - Resultado de la función lineal.
    """
    return 17 * x - 210.3333

def g(x):
    """
    Calcula el valor de la función exponencial g(x) = 0.1452 * e^(0.369x).
    @param {float or np.array} x - Valor(es) de x donde se evaluará la función.
    @return {float or np.array} - Resultado de la función exponencial.
    """
    return 0.1452 * np.exp(0.369 * x)

def h(x):
    """
    Calcula el valor de la función logarítmica h(x) = 259.49 * ln(x) - 656.46.
    @param {float or np.array} x - Valor(es) de x donde se evaluará la función.
    @return {float or np.array} - Resultado de la función logarítmica.
    """
    return 259.49 * np.log(x) - 656.46

# Rango de valores en el dominio [13, 18]
x_vals = np.linspace(13, 18, 100)

# Figura
plt.figure(figsize=(10, 6))
# Puntos de dispersión
plt.scatter(x, y, color="blue", label="Puntos de dispersión")
# Función lineal
plt.plot(x_vals, f(x_vals), color="red", label="$f(x) = 17x - 210.3333$")
# Función exponencial
plt.plot(x_vals, g(x_vals), color="green", label="$g(x) = 0.1452e^{0.369x}$")
# Función logarítmica
plt.plot(
    x_vals,
    h(x_vals),
    color="purple",
    linestyle=":",
    label="$h(x) = 259.49 ln(x) - 656.46$",
)

# Configuración general del gráfico
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfico 1")
plt.legend()
plt.grid(True)
plt.show()
```

##### **Observaciones Gráfico 1:**
- **Crecimiento Diferenciado:** La función lineal *f(x)* muestra un crecimiento constante, la función exponencial *g(x)* aumenta más rápido a medida que avanza *x*, y la función logarítmica *h(x)* crece rápido al inicio pero se estabiliza.

```python
import pandas as pd

# Ruta fichero
file_path = "Anexo 1 - Base de Datos Excel.xlsx"

# Carga del DataFrame - hoja "BASE DE DATOS".
df1 = pd.read_excel(
    file_path,
    sheet_name="BASE DE DATOS",
    header=3,
    index_col=0,  # Usar la primera columna como índice
)

df1.head()
```

```python
import matplotlib.pyplot as plt

# Nombre de las columnas en el Dataframe df1
col_time_motor_on = 'Tiempo motor ON (h)'
col_time_motion = 'Tiempo en Movimiento (h)'

#Figura 1 histogramas
plt.figure(figsize=(12, 6))

# Histogramas
plt.hist(df1[col_time_motor_on], bins=20, alpha=0.5, color='gray', label='Tiempo motor ON')
plt.hist(df1[col_time_motion], bins=20, alpha=0.5, color='green', label='Tiempo en Movimiento')

# Configuración general del gráfico
plt.title('Histograma: Tiempo motor ON vs Tiempo en Movimiento')
plt.xlabel('Tiempo (horas)')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Figura 2 gráfico de dispersión
plt.figure(figsize=(12, 6))

# Gráfico de dispersión
plt.scatter(df1[col_time_motion], df1[col_time_motor_on], color='blue', alpha=0.7)

# Configuración general del gráfico
plt.title('Gráfico de Dispersión: Tiempo en Movimiento vs Tiempo motor ON')
plt.xlabel('Tiempo en Movimiento (horas)')
plt.ylabel('Tiempo motor ON (horas)')
plt.grid(True)
plt.show()
```
